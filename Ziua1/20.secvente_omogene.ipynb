# Range - colectie de intregi

```python
for i in range(3):
    print("i=", i)
```

```python
for i in range(32, 35):
    print("i=", i)
```

```python
for i in range(100, 90):
    print("i=", i)
```

```python
for i in range(100, 90, -1):
    print("i=", i)
```

```python
for i in range(100, 90, -3):
    print("i=", i, type(i))
```
