```python
with open("numere.txt", "r") as file_handler:
    continut = file_handler.read()
    print("continut:", continut)
```

```python
def suma_numerelor_unice(numere):
    numere = [int(i) for i in numere]
    numere_unice = [i for i in numere if numere.count(i) == 1]
    return sum(numere_unice)
```

```python
lista = continut.split(",")

suma_numerelor_unice(lista)
```
