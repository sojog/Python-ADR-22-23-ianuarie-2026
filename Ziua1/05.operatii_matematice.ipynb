# Float si int sunt compatibile

```python
# Adunare
print(1.3 + 12)
# Scadere
print(23.3 - 23)
# Inmultire
print(2 * 7.2)
# Impartire
print( 10.2 / 5)
```

# Bool, float si int sunt compatibile (matematic)

```python
# Adunare
print(1.3 + 12 + True) # True este convertit la 1
# Scadere
print(23.3 - 23 + True)
# Inmultire
print(2 * 7.2  + True)
# Impartire
print( 10.2 / 5 + True)
```

```python
# Adunare
print(1.3 + 12 + False) # True este convertit la 0
# Scadere
print(23.3 - 23 + True)
# Inmultire
print(2 * 7.2  + True)
# Impartire
print( 10.2 / 5 + True)
```

# Operatiile matematice nu pot fi facute pe string (*)

```python
# Adunare
print(1.3 + 12 + "312")
```

```python
# Adunare
print(1.3 + 12 + int("312"))
```
