```python
def adunare(a, b):
    return a + b

def scadere(a, b):
    return a - b

def inmultire(a, b):
    return a * b
```

```python
def executa_operatia(functia, a, b):
    return functia(a, b)
```

```python
executa_operatia(adunare, 10, 2)
```

```python
executa_operatia(scadere, 9, 6)
```

```python
executa_operatia(inmultire, 3, 5)
```
