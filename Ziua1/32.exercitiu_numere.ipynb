```python
with open("numere.txt", "r") as file_handler:
    continut = file_handler.read()
    print("continut:", continut)
```

```python
numere_stringuri = continut.split(",")
numere_stringuri
```

```python
numere = []
for i in numere_stringuri:
    numere.append(int(i))
numere
```

```python
# List comprehention
numere = [int(i) for i in numere_stringuri]
numere
```

```python
intalniri = { }
for i in numere:
    print("i", i)
    intalniri[i] = intalniri.get(i, 0) + 1
intalniri
```

```python
suma = 0
for i in intalniri:
    if intalniri[i] == 1:
        suma += i
print(suma)
```
