```python
def bancnote(suma, bancnote = [10, 5, 2, 1]):
    rezultat = {}
    for b in bancnote:
        nr = suma // b
        if nr > 0:
            # print(f"{nr} bancnote de {b}")
            rezultat[b] = nr 
            suma = suma % b
    return rezultat

bancnote(128)
```

```python
def bancnote(suma, bancnote = [10, 5, 2, 1]):
    bancnote = sorted(bancnote, reverse=True)
    rezultat = {}
    for b in bancnote:
        nr = suma // b
        if nr > 0:
            rezultat[b] = nr 
            suma = suma % b
    return rezultat

bancnote(128, [20, 10, 100, 1 ])
```
