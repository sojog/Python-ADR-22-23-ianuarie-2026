```python
bancnote = [ 10, 5, 2, 1]

suma = int(input("Introduceti o suma: "))

for b in bancnote:
    nr = int(suma / b)
    print(f"Nr bancnote de {b}:", nr)
    suma = suma / b
```

```python
suma = int(input("Introdu suma: "))

bancnote = [10, 5, 2, 1]

for b in bancnote:
    nr = suma // b
    if nr > 0:
        print(f"{nr} bancnote de {b}")
        suma = suma % b
```
