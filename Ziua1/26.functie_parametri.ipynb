```python
def suma_totala(*params):
    print(params, type(params))
```

```python
suma_totala()
```

```python
suma_totala(2)
```

```python
suma_totala(10, 230, 320, 304, 503, 32)
```

```python
def functie_cu_parametri(x, y, *params):
    print(params, type(params))
```

```python
functie_cu_parametri(10, 230, 320, 304, 503, 32)
```

```python
def functie_cu_parametri_optionali( *args, **kargs):
    print(args, type(args))
    print(kargs, type(kargs))
```

```python
functie_cu_parametri_optionali(10, 20, 30)
```

```python
functie_cu_parametri_optionali(10, 20, 30, cheie="valoare", alta_cheie=31231)
```
