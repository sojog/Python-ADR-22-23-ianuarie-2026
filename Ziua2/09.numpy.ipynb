# 2. Filtrare

```python
arr = np.array([11, 22, 33, 44, 55])
arr > 22
```

```python
arr[arr > 22]
```

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77, 88, 99, 100])

arr[(arr > 22) & (arr < 88)] # AND -> 2 conditii simulan
```

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77, 88, 99, 100])

arr[(arr <= 22) | (arr >= 88)] # OR -> 2 conditii alternativ
```

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77, 88, 99, 100])

conditia1 = (arr <= 22) 
conditia2 = (arr >= 88)

arr[conditia1 | conditia2 ] # OR -> 2 conditii alternativ
```
