```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
print(plt.__doc__)
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = 2
y = 3
plt.scatter(x, y)
plt.show()
```

```python
x = 2, 4
y = 3, 5
plt.scatter(x, y)
plt.show()
```

```python
x = 2, 4
y = 3, 5
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
x = (2, 4, 7)
y = (3, 5, 6)
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
x = (2, 4, 7)
y = (3, 5, 6)
plt.scatter(x, y, c="red")
plt.plot(x, y)
plt.show()
```

```python
x = (2, 4, 7)
y = (3, 5, 6)
plt.scatter(x, y, c="red")
plt.plot(x, y)
plt.xlabel("Coordonata x")
plt.ylabel("Coordonata y")
plt.show()
```

```python
x = (2, 4, 7)
y = (3, 5, 6)
plt.bar(x, y)
plt.show()
```

```python
x = (2, 4, 7)
y = (3, 5, 6)
plt.barh(x, y)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.pie(x, labels=["A", "B", "C"])
plt.show()
```
