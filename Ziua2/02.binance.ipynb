```python
import requests
```

```python
BASE_URL = "https://api1.binance.com"
AVG_PRICE_PATH = "/api/v3/avgPrice"


URL = BASE_URL + AVG_PRICE_PATH
URL
```

```python
params = {
    "symbol" : "BTCUSDT"
}
```

```python
response = requests.get(URL,params=params)
```
