```python
import pandas as pd
```

```python
URL = "https://raw.githubusercontent.com/sojog/Python-ADR-22-23-ianuarie-2026/refs/heads/main/Ziua2/clwin.csv"
```

```python
df = pd.read_csv(URL, index_col=0)
df
```

```python
teams = [["Spain", "Real", 15], ["Italy", "Milan", 7], ["Germany", "Bayern", 6]]
```

```python
big_clubs_df = pd.DataFrame(teams, columns=["country", "team", "trophies"])
big_clubs_df
```

```python
merged_df = pd.concat([df, big_clubs_df ])
merged_df
```

```python
merged_df = merged_df.reset_index(drop=True)
merged_df
```

```python
merged_df
```

```python
merged_df.sort_values("trophies", ascending=False)
```

```python
merged_df = merged_df.sort_values(["trophies", "team"], ascending=[False, True]).reset_index(drop=True)
merged_df
```
