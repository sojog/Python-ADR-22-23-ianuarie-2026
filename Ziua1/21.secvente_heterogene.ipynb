# Tuple - colectie heterogena, imutabila

```python
x = [10, 200, 50, 41, 52]
print(x, type(x))
```

```python
x = (10, 200, 50, 41, 52)
print(x, type(x))
```

```python
x = (10, 200, 50, 41, 52)
for i in x:
    print("i=", i)
```

```python
x = (10, 200, 50, 41, 52)
print(len(x))
```

```python
x = (10, 200, 50, 41, 52)
print(x[-2:])
```
