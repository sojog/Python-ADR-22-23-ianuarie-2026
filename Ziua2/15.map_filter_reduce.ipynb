```python
culori = ['rosu', 'verde', 'albastru']
```

```python
def ascunde(text):
    return len(text) * "*"
```

```python
list(map(   ascunde     , culori))
```

```python
import pandas as pd
```

```python
culori = ['rosu', 'verde', 'albastru']
temperaturi = [1, 2, 3]

df = pd.DataFrame({"culori": culori, "temperaturi": temperaturi})
df
```

```python

df["culori"] = df["culori"].map(ascunde)
df
```
