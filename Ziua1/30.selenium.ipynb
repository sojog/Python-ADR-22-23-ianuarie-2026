```python
from selenium import webdriver


driver = webdriver.Chrome()
```

```python
driver.get('https://www.cursbnr.ro/')
```

```python
from selenium.webdriver.common.by import By
```

```python
x_path_consent = "/html/body/div[6]/div[2]/div[2]/div[2]/div[2]/button[1]/p"
```

```python
button_consintamanat = driver.find_element(By.XPATH, x_path_consent)
button_consintamanat
```

```python
button_consintamanat.click()
```
