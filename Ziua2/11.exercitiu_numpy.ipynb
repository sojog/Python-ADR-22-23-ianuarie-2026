```python
import numpy as np

arr = np.arange(3000)

print (arr[arr%4==0])
```

```python
# Este bisect daca
# conditia1: este divizibil cu 400
#                        | (sau)
# conditia2: este divizibil cu 4 & (dar nu cu 100)
```

```python
arr = np.arange(100)

print (arr[arr%4 != 0])
```

```python
start_date = 1899
stop_date = 2126

arr = np.arange(start_date, stop_date)

conditia1 = arr % 400 == 0
conditia2 = (arr % 4 == 0) & (arr % 100 != 0)

arr[conditia1 | conditia2]
```

```python
start_date = 1899
stop_date = 2126

arr = np.arange(start_date, stop_date)

arr[(arr % 400 == 0) | ((arr % 4 == 0) & (arr % 100 != 0))]
```
