# AND, OR, NOT

```python
# AND
print( True and True )
print( True and False )
print( False and True )
print( False and False )
```

```python
# OR
print( True or True )
print( True or False )
print( False or True )
print( False or False )
```

```python
# NOT
print( not True )
print( not False )
```

# Operatii logice pe alte tipuri de valori decat bool

```python
print( 12 and 165 )
print( 23 and 0 )
print( 0 and 321 )
print( 0 and 0 )
```
