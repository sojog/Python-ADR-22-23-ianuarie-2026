# Fisiere text

```python
# r - read
# w - write (rescriere)
# a - append (adaugare)
```

```python
# Versiunea 1 - mai veche
file_handler = open("temperaturi.txt", "a")
file_handler.write("10, 20, 30, 30")
file_handler.close()
```

```python
# Versiunea 2 - preferata
with open("temperaturi.txt", "w") as file_handler:
    file_handler.write("10, 20, 30, 30")
```

```python
with open("temperaturi.txt", "r") as file_handler:
    continut = file_handler.read()
    print("continut:", continut)
```
